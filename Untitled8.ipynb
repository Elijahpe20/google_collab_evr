```python
distancia_km = 350
consumo_por_km = 0.08
precio_litro = 1.55

litros_necesarios = distancia_km * consumo_por_km
costo_total = litros_necesarios * precio_litro

km_sobrantes = distancia_km % 100

print(f"El costo total del viaje es: ${costo_total}")
print(f"Después de cada parada de 100km, sobran {km_sobrantes} km.")
```

```python
amigos_Elias = {"Lolo", "Cynthia", "Marly", "Junior"}
amigos_Julio = {"Marisol", "Junior", "Felipe", "Marfil"}

todos_los_invitados = amigos_Elias.union(amigos_Julio)
print(todos_los_invitados)

amigos_comunes = amigos_Elias.intersection(amigos_Julio)
print(amigos_comunes)

solo_Elías = amigos_Elias.difference(amigos_Julio)
print(solo_Elías)

solo_Julio = amigos_Julio.difference(amigos_Elias)
print(solo_Julio)
```
