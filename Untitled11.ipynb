```python
#1 ejercicio: ⭐ Calcula P(default=1).
import pandas as pd

# 1. Cargar la base de datos
df = pd.read_csv('base_datos_ejercicios_completa.csv')

# 2. Calcular la probabilidad
# La media de una columna binaria (0 y 1) es igual a la probabilidad de que sea 1.
p_default_1 = df['default'].mean()

# 3. Mostrar el resultado
print(f"La probabilidad P(default = 1) es: {p_default_1:.4f}")
print(f"En porcentaje: {p_default_1 * 100:.2f}%")
```

Explicación Teórica y AnálisisEl resultado en tu base de datos es 13.75% (0.1375).¿Qué significa esto para un analista financiero?Contraintuitivo: Curiosamente, en estos datos específicos, la probabilidad de default para alguien con 3 o más pagos tarde (13.75%) es menor que la probabilidad general de toda la base (17.53%).Lección Clave: Esto suele suceder en datasets de ejercicios o en casos reales donde existen otras variables "protectoras".
```python
#5 ejercicio: Verifica la fórmula P(B ∪ C)=P(B)+P(C)-P(B ∩ C) para B=(dti>0.35) y C=(pagos_tarde_12m>=3).

# Definimos los eventos lógicos
B = df['dti'] > 0.35
C = df['pagos_tarde_12m'] >= 3

# 1. Calculamos las probabilidades individuales
p_b = B.mean()
p_c = C.mean()

# 2. Calculamos la Intersección (B y C ocurren al mismo tiempo)
# Usamos el símbolo '&' para representar la intersección (AND)
p_interseccion = (B & C).mean()

# 3. Calculamos la Unión Real (B o C ocurren)
# Usamos el símbolo '|' para representar la unión (OR)
p_union_real = (B | C).mean()

# 4. Aplicamos la fórmula teórica para verificar
p_union_teorica = p_b + p_c - p_interseccion

print(f"P(B) [dti > 0.35]: {p_b:.4f}")
print(f"P(C) [pagos_tarde >= 3]: {p_c:.4f}")
print(f"P(B ∩ C) [Intersección]: {p_interseccion:.4f}")
print("-" * 30)
print(f"P(B ∪ C) Medido directamente: {p_union_real:.4f}")
print(f"P(B ∪ C) Calculado por fórmula: {p_union_teorica:.4f}")
```

2. Explicación del Código y los Operadores¿Qué hace el & (Ampersand)?Es el operador de intersección. En Python, cuando comparas dos series de True/False con &, solo resultará True si ambas son verdaderas. Es decir, clientes que tienen mucha deuda Y ADEMÁS pagan tarde.¿Qué hace el | (Pipe)?Es el operador de unión. Resultará True si el cliente cumple al menos una de las condiciones: o tiene mucha deuda, o paga tarde, o ambas.

```python
#7 ejercicio: ⭐ Define un “alerta de riesgo” con sensibilidad=0.8 y especificidad=0.75. Con la tasa base de tu dataset, calcula P(default | alerta=1).
# --- EJERCICIO 6: TEOREMA DE BAYES ---

# 1. Tasa base de tu dataset (P(default=1))
p_default = df['default'].mean()
p_no_default = 1 - p_default

# 2. Características de la "Alerta de Riesgo" (Dadas por el ejercicio)
sensibilidad = 0.80  # P(alerta=1 | default=1) -> Detecta al 80% de los malos.
especificidad = 0.75 # P(alerta=0 | default=0) -> Detecta al 75% de los buenos.
tasa_falsos_positivos = 1 - especificidad # P(alerta=1 | default=0) -> 25% error.

# 3. Probabilidad total de que la alerta se encienda P(alerta=1)
# Puede encenderse porque el cliente es malo (cierto) o por error (falso positivo)
p_alerta = (sensibilidad * p_default) + (tasa_falsos_positivos * p_no_default)

# 4. Aplicamos Bayes: P(default=1 | alerta=1)
p_final_bayes = (sensibilidad * p_default) / p_alerta

print(f"Probabilidad Inicial (Sin Alerta): {p_default*100:.2f}%")
print(f"Probabilidad Final (Si la Alerta=1): {p_final_bayes*100:.2f}%")
```

```python
#8 ejercicio: ⭐ Repite el ejercicio 7 cambiando la tasa base a 2% (0.02). ¿Qué ocurre? (problema de tasa base)
import pandas as pd

# 1. Cargamos el archivo
df = pd.read_csv('base_datos_ejercicios_completa.csv')

# 2. Contamos cuántos hay de cada uno en la columna 'default'
conteo = df['default'].value_counts()

print("--- Conteo Real en tu base de datos ---")
print(f"Clientes Buenos (0): {conteo[0]}")
print(f"Clientes Malos (1): {conteo[1]}")
print(f"Total: {conteo[0] + conteo[1]}")

# 3. Calculamos la probabilidad (tasa base)
tasa_base = conteo[1] / len(df)
print(f"\nProbabilidad inicial: {tasa_base:.4f} ({tasa_base*100:.2f}%)")
```

Este ejercicio es brillante porque te muestra el "Peligro de la Tasa Base" (o Falacia de la Tasa Base), un fenómeno que confunde incluso a profesionales expertos.

Aquí tienes la resolución y la explicación de por qué los números se comportaron de forma tan extraña.

1. El Resultado del Cálculo (Bayes con 2%)
Si cambiamos la probabilidad inicial de default al 2%, pero mantenemos la misma alerta (80% sensibilidad, 75% especificidad), el resultado es:

Tasa Base Inicial: 2.00%

Probabilidad si suena la alerta (P(D∣A)): 6.13%

2. ¿Qué ocurrió? (La explicación sencilla)
En el ejercicio anterior (con 17.5% de tasa base), la alerta hacía que el riesgo subiera al 40%. Ahora, con la misma alerta, el riesgo solo sube al 6%.

¿Por qué la alerta parece "peor" ahora? No es que la alerta haya cambiado, es que el problema es mucho más raro.

Usemos de nuevo el ejemplo de la habitación para que veas la "trampa":

Imagina 1,000 personas. Como la tasa es del 2%, solo hay 20 malos y 980 buenos.

La Alerta actúa:

Atrapa al 80% de los malos: 16 personas. (Aciertos)

Se equivoca con el 25% de los buenos: 245 personas. (Errores/Falsos Positivos)

La Habitación de Sospechosos:

Ahora tienes una habitación con 16+245=261 personas.

La Realidad:

Aunque la alerta sonó, la habitación está inundada de gente buena (245) y solo hay 16 malos.

Por eso, la probabilidad de que alguien ahí sea malo es de solo 16/261=6.13%.

3. La gran lección: El problema de la Tasa Base
Esta es la conclusión que debes aprender para tu carrera:

"Cuando un evento es muy raro (tasa base baja), incluso una prueba muy buena dará más resultados falsos que verdaderos".

¿Cómo se aplica esto a las finanzas?
Si un banco lanza una campaña para un producto muy seguro donde casi nadie falla (tasa base del 2%), y usa una alerta de riesgo que se equivoca el 25% de las veces:

El banco va a estar rechazando a muchísima gente buena por error.

De cada 100 alertas que suenen, 94 serán clientes buenos marcados injustamente y solo 6 serán deudores reales.

Moraleja: Si el problema que buscas es muy escaso (como el fraude o el default en clientes VIP), necesitas una alerta extremadamente precisa (especificidad de 99% o más), porque si no, los errores "sepultan" a los aciertos.

¿Te ha volado la cabeza este resultado? Es la razón por la que Bayes es tan respetado: nos obliga a ser humildes con nuestras "alertas" e "intuiciones".

```python
#8 ejercicio: ⭐ Si un analista cree que P(default)=0.10 en un segmento, pero ve una señal con P(señal|default)=0.7 y P(señal|no default)=0.2, calcula P(default|señal).

# --- EJERCICIO: BAYES EN UN SEGMENTO ESPECÍFICO ---

# 1. Creencia inicial (Prior)
p_default = 0.10
p_no_default = 1 - p_default # 0.90

# 2. Datos de la señal (Likelihoods)
# P(señal | default) es la Sensibilidad
p_senal_dado_default = 0.70

# P(señal | no default) es la Tasa de Falsos Positivos
p_senal_dado_no_default = 0.20

# 3. Probabilidad total de ver la señal P(señal)
# (Aciertos + Errores)
p_total_senal = (p_senal_dado_default * p_default) + (p_senal_dado_no_default * p_no_default)

# 4. Cálculo final de Bayes: P(default | señal)
p_posterior = (p_senal_dado_default * p_default) / p_total_senal

print(f"Riesgo inicial: {p_default*100:.2f}%")
print(f"Riesgo tras ver la señal: {p_posterior*100:.2f}%")
```

```python
#16  ⭐ Calcula z-score para un umbral de log(ingreso) y estima P(X>umbral) usando la CDF normal.
import pandas as pd
import numpy as np
from scipy.stats import norm

# 1. Cargar datos y transformar a logaritmo
df = pd.read_csv('base_datos_ejercicios_completa.csv')
df['log_ingresos'] = np.log(df['ingresos_anuales'])

# 2. Calcular estadísticos
mu = df['log_ingresos'].mean()
sigma = df['log_ingresos'].std()

# 3. Definir un umbral (ejemplo: $70,000) y calcular su Z-score
umbral_dinero = 70000
log_umbral = np.log(umbral_dinero)

z_score = (log_umbral - mu) / sigma

# 4. Calcular probabilidad P(X > umbral) usando la CDF (área bajo la curva)
# norm.cdf(z) da la probabilidad a la izquierda, por eso restamos de 1
probabilidad_mayor = 1 - norm.cdf(z_score)

print(f"Z-score para ${umbral_dinero}: {z_score:.4f}")
print(f"Probabilidad de ganar más de ${umbral_dinero}: {probabilidad_mayor:.4f}")
```

Este código calcula qué tan lejos está un sueldo de la media y la probabilidad de encontrar a alguien con ese sueldo o más.

```python
#17 Calcula qué % de registros tienen |z|>3 en log(ingreso).
# 1. Calcular el Z-score para cada fila de la tabla
df['z_score'] = (df['log_ingresos'] - mu) / sigma

# 2. Filtrar los registros que están a más de 3 desviaciones estándar (absoluto)
outliers = df[df['z_score'].abs() > 3]

# 3. Calcular el porcentaje
porcentaje_outliers = (len(outliers) / len(df)) * 100

print(f"Número de outliers detectados: {len(outliers)}")
print(f"Porcentaje de la base de datos: {porcentaje_outliers:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# 1. Cargar el dataset
df = pd.read_csv('base_datos_ejercicios_completa.csv')

# --- EJERCICIO 26: Regla de Riesgo Multivariable ---
def evaluar_regla(df, condicion):
    predicho = condicion.astype(int)
    real = df['default']
    tp = ((predicho == 1) & (real == 1)).sum()
    fp = ((predicho == 1) & (real == 0)).sum()
    tn = ((predicho == 0) & (real == 0)).sum()
    fn = ((predicho == 0) & (real == 1)).sum()
    return {"TP": tp, "FP": fp, "TN": tn, "FN": fn}

# Comparar Regla Simple (DTI) vs Regla Compuesta (DTI + Pagos Tarde)
regla_original = df['dti'] > 0.35
regla_nueva = (df['dti'] > 0.35) | (df['pagos_tarde_12m'] >= 2)

res_orig = evaluar_regla(df, regla_original)
res_nueva = evaluar_regla(df, regla_nueva)

print("--- EJERCICIO 26: COMPARACIÓN ---")
print(f"Original: {res_orig}")
print(f"Multivariable: {res_nueva}")

# --- EJERCICIO 27 & 28: Funciones 'Al menos k' ---
def binom_cdf_ge(k, n, p):
    """P(K >= k) para Binomial"""
    return 1 - binom.cdf(k - 1, n, p)

def poisson_cdf_ge(k, lam):
    """P(K >= k) para Poisson"""
    return 1 - poisson.cdf(k - 1, lam)

# --- EJERCICIO 29: Segmentación por Bins de DTI ---
df['dti_bins'] = pd.cut(df['dti'], bins=np.arange(0, 1.1, 0.1))
dti_risk = df.groupby('dti_bins', observed=False)['default'].mean()

plt.figure(figsize=(10, 6))
dti_risk.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Riesgo de Default por Segmento de DTI')
plt.ylabel('Probabilidad de Default')
plt.savefig('riesgo_por_segmentos.png')

# --- EJERCICIO 30: Mini-Reporte de Aprendizaje ---
print("\n--- EJERCICIO 30: REPORTE DE INCERTIDUMBRE ---")
print("""
1. La incertidumbre es cuantificable: el riesgo no es 'azar', es una probabilidad medible.
2. Los errores tienen precio: un Falso Negativo (FN) es mucho más caro que un Falso Positivo (FP).
3. La tasa base es el ancla: ignorar la frecuencia histórica lleva a predicciones erróneas.
4. Bayes permite aprender: una señal nueva actualiza nuestra creencia previa con precisión.
5. Segmentar revela la verdad: los promedios ocultan riesgos en nichos específicos.
6. Multivariable es mejor: combinar señales (DTI + Pagos) reduce las pérdidas drásticamente.
7. El tamaño importa: las muestras grandes estabilizan las métricas y dan confianza.
8. El logaritmo normaliza: transformar datos ayuda a que los modelos entiendan mejor el dinero.
9. El umbral es una decisión: ser estricto protege el capital pero puede alejar clientes.
10. La estadística es defensa: permite al banco sobrevivir a escenarios de mucho estrés.
""")
```
