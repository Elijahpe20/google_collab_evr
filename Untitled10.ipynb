```python
import numpy as np

n = int(input("¿Cuántos productos vas a registrar?: "))

# Creamos listas vacías para guardar los datos temporalmente
lista_precios = []
lista_cantidades = []

for i in range(n):
    print(f"--- Producto {i+1} ---")
    p = float(input("Precio unitario: "))
    c = int(input("Cantidad en stock: "))

    lista_precios.append(p)
    lista_cantidades.append(c)

# Convertimos las listas a Arreglos Numpy
arr_precios = np.array(lista_precios)
arr_cantidades = np.array(lista_cantidades)

# AQUÍ ESTÁ LA MAGIA DE NUMPY:
# Multiplicamos dos arrays directamente sin usar otro bucle for
arr_subtotales = arr_precios * arr_cantidades

print("\nResultados:")
print("Precios:", arr_precios)
print("Cantidades:", arr_cantidades)
print("Subtotales por producto:", arr_subtotales)
print(f"--> VALOR TOTAL DEL INVENTARIO: {arr_subtotales.sum()}")
```

```python
import numpy as np
dias =int(input("¿Cuántos días desea registrar?: "))
temps = np.zeros(dias)

for i in range(dias):
    temps[i] = float(input(f"Ingrese temperatura del día {i+1}: "))

promedio = np.mean(temps)
dias_calurosos = temps[temps > 25]

print("\n--- Reporte Climático Final---")
print(f"Temperatura Promedio: {promedio:.2f}°C")
print(f"Cantidad de días calurosos (>25°): {len(dias_calurosos)}")

if len(dias_calurosos) > 0:
  print(f"Las temperaturas calurosas fueron: {', '.join(map(str, dias_calurosos))}")
else:
  print("No se registraron temperaturas calurosas.")
```

```python
import numpy as np
n_ventas = int(input("Ingrese cantidad de ventas a registrar: "))
montos = []
vendedores = []

for i in range(n_ventas):
    m = float(input(f"Venta {i+1} - Monto $: "))

    while True:

        v =int(input(f"Venta {i+1} - Vendedor (1=Ana, 2=Juan: )"))
        if v == 1 or v == 2:
          break
        print("Error: Ingrese 1 o 2")
    montos.append(m)
    vendedores.append(v)

arr_montos = np.array(montos)
arr_vendedores = np.array(vendedores)

ventas_ana = arr_montos[arr_vendedores == 1]
ventas_juan = arr_montos[arr_vendedores == 2]

total_ana = ventas_ana.sum()
total_juan = ventas_juan.sum()

print("\n--- Resultados de la COmpetencia ---")
print(f"Total vendido por Ana: ${total_ana}")
print(f"Total vendido por Juan: ${total_juan}")

if total_ana > total_juan:
  print ("¡GANADORA_ ANA!")
elif total_juan > total_ana:
  print("¡GANADOR_JUAN!")
else:
  print("¡EMPATE!")
```
