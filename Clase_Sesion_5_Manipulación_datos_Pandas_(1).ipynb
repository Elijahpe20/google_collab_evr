# Aplicar operaciones entre columnas

```python
# agregamos en una nueva columna el ratio de habitantes por superficie
df['Habit_x_km2'] = (df['Poblacion'] / df['Superficie']).astype(int)
df.sort_values(['Habit_x_km2'])
```

## Aplicar una operacion definida

```python
def crear_codigo(name):
    name = name.upper() # paso a mayusculas
    name = name[0:4]    # primeros 4 char
    return name
```

### Aplicamos usando 1 columna

```python
df['Codigo'] = df['Pais'].apply(crear_codigo)
df
```

### Aplicamos una funci√≥n enviando cada fila (todas las columnas)

```python
def categoria(fila):
    pob = fila['Poblacion']
    habit = fila['Habit_x_km2']
    if pob > 46000000:
        if habit <50:
            return 'A'
        else:
            return 'B'

    return 'C'

df['Categoria'] = df.apply(categoria, axis=1)
df
```

### Aplicar enviando algunas columnas como par√°metros

```python
def asigna_color(codigo, categoria):
    if categoria=='A':
        return 'rojo'
    if codigo == 'ESPA':
        return 'verde'
    return 'azul'

df['color'] = df.apply(lambda x: asigna_color(x['Codigo'],x['Categoria']), axis=1);
df
```

### Mapeo

```python
df['mapeo_color'] = df['color'].map( {'azul': 0, 'rojo': 1, 'verde': 2} ).astype(int)
df
```

# Reordenamos columnas

## Left join por columna clave (merge)

```python
# Imaginemos que tenemos un DF nuevo, le asignamos el Codigo como identificador √∫nico
df_factor = pd.DataFrame(data={"Categoria" : ["A","B","C"],
                            "Factor":[12.5,103,0.001]}
                        )
df_factor
```

```python
result2 = pd.merge(df, df_factor, how='left', on=['Categoria'])
result2
```

# Preguntas

0.  **Data Generation**: Create a Pandas DataFrame named `df_exam` with fictional data including columns for 'City', 'Country', 'Region', 'Population', and 'Surface Area'.
1.  **Question 1 (Basic Exploration and Filtering)**:
    *   Display the first 5 rows of `df_exam`.
    *   Filter and display cities located in 'South America' with a 'Population' greater than 10,000,000.
2.  **Question 2 (New Column Creation and Sorting)**:
    *   Calculate a new column 'Population_Density' (Population / Surface Area) and add it to `df_exam`.
    *   Sort `df_exam` by 'Population_Density' in descending order and display the result.
3.  **Question 3 (Grouping and Aggregation)**:
    *   Group `df_exam` by 'Region' and calculate the total 'Population' for each region.
    *   Group `df_exam` by 'Country' and find the average 'Surface Area' for each country.
4.  **Question 4 (Row/Column Manipulation and Update)**:
    *   Remove the 'Population_Density' column from `df_exam`.
    *   Add a new city 'Lima', 'Peru', 'South America', with a population of 10,500,000 and a surface area of 3500 to `df_exam`.
    *   Update the 'Population' of 'Mexico City' to 22,000,000 in `df_exam`.
5.  **Question 5 (Data Visualization)**:
    *   Create a bar chart showing the 'Population' of the top 5 cities (by population).
    *   Create a pie chart showing the 'Surface Area' distribution by 'Country'.

Provide the Python code solutions for each question.

```python
## Respuesta Pregunta 0 ##

import pandas as pd

data = {
    'City': ['Buenos Aires', 'Bogota', 'Madrid', 'Mexico City', 'Caracas', 'Santiago'],
    'Country': ['Argentina', 'Colombia', 'Spain', 'Mexico', 'Venezuela', 'Chile'],
    'Region': ['South America', 'South America', 'Europe', 'North America', 'South America', 'South America'],
    'Population': [15500000, 11500000, 6700000, 23000000, 2900000, 7000000],
    'Surface Area': [4000, 1775, 604, 1485, 433, 641]
}

df_exam = pd.DataFrame(data)
print("DataFrame 'df_exam' created successfully:")
display(df_exam)
```

```python
## Respuesta Pregunta 01 ##
# Parte A: Mostrar solo las primeras 5 filas
print("--- A. Primeras 5 filas ---")
display(df_exam.head())

# Parte B: Filtrar (Sudam√©rica Y Poblaci√≥n > 10 millones)
print("\n--- B. Ciudades grandes de Sudam√©rica ---")
condicion = (df_exam['Region'] == 'South America') & (df_exam['Population'] > 10000000)
resultado_filtro = df_exam[condicion]

display(resultado_filtro)
```

```python
## Respuesta Pregunta 02 ##
df_exam['Population_Density'] = df_exam['Population'] / df_exam['Surface Area']
df_ordenado = df_exam.sort_values(by='Population_Density', ascending=False)

print("--- Tabla ordenada por Densidad ---")
display(df_ordenado)
```

```python
## Respuesta Pregunta 03 ##
poblacion_por_region = df_exam.groupby('Region')['Population'].sum()
print("--- Poblaci√≥n Total por Regi√≥n ---")
display(poblacion_por_region)
area_promedio_pais = df_exam.groupby('Country')['Surface Area'].mean()
print("\n--- √Årea Promedio por Pa√≠s ---")
display(area_promedio_pais)
```

```python
## Respuesta Pregunta 04 ##
# 1. Eliminar la columna de densidad
# 'axis=1' significa que borramos una columna (vertical), no una fila.
df_exam = df_exam.drop('Population_Density', axis=1)

# 2. Agregar la nueva fila (Lima, Peru...)
# Creamos un DataFrame peque√±ito con la nueva fila
nueva_fila = pd.DataFrame({
    'City': ['Lima'], 'Country': ['Peru'], 'Region': ['South America'],
    'Population': [10500000], 'Surface Area': [3500]
})
# Lo pegamos al original
df_exam = pd.concat([df_exam, nueva_fila], ignore_index=True)

# 3. Actualizar la poblaci√≥n de Mexico City
# Le decimos: "Busca donde City sea Mexico City, y en la columna Population pon este valor"
df_exam.loc[df_exam['City'] == 'Mexico City', 'Population'] = 22000000

print("--- Tabla actualizada con Lima y cambios en M√©xico ---")
display(df_exam.tail()) # tail() muestra el final, donde debe salir Lima
```

```python
## Respuesta Pregunta 05 ##
# Necesitamos importar matplotlib para que los gr√°ficos se vean bien
import matplotlib.pyplot as plt

# 1. Gr√°fico de Barras: Top 5 ciudades por poblaci√≥n
# Primero ordenamos las 5 m√°s grandes (nlargest) y luego graficamos
top_5 = df_exam.nlargest(5, 'Population')
top_5.plot(kind='bar', x='City', y='Population', title='Top 5 Ciudades por Poblaci√≥n', color='skyblue')
plt.show()

# 2. Gr√°fico de Pastel: Distribuci√≥n de √Årea por Pa√≠s
# Usamos el Pa√≠s como √≠ndice para que aparezcan los nombres en el pastel
df_exam.set_index('Country').plot(kind='pie', y='Surface Area', title='Distribuci√≥n de √Årea', legend=False, figsize=(6,6))
plt.ylabel('') # Esto es est√©tico, para quitar la etiqueta del eje Y
plt.show()
```
